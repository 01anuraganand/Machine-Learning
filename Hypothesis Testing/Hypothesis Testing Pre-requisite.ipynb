# Hypothesis Testing 
 <h2>Test of Significance</h2> Test of significance enables us to decide on the basis of sample results that whether the deviation  between the observed value (statistics) and hypothetical parameter or deviation between two independent statistics is significant or merly due to fluctuations of samples
 
 <h2> Hypothesis</h2>
 It is statement made about the parameter of population or difference between the value of observed statistics and hypothetical parameters.
<br>

Two types of Hypothesis-


<b>1. Null Hypothesis</b>- Null hypothesis is hypothesis which is tested for the possible rejection under the assumption that it  is True. It is denoted by $H_o : \mu  = \mu_o $ ,
where $\mu$ - parameter  of population and $\mu_o$ - corresponding sample statistics
<br>

<b>2. Alternate Hypothesis</b> - Any hypothesis which is complementry to Null Hupothesis is called Alternate Hypothesis. It is denoted by $H_1  or  H_a$

$H_1 : \mu \neq \mu_o$ ( Two Tailed)

$H_1 : \mu > \mu_o$ (Right Tailed)

$H_2 : \mu < \mu_o$ (Left Tailed)

<h2> Error In Sampling</h2>

The main objective of sample theory is to draw right information about the population parameter on the basis of sample result.
In practice we may decide to accept or reject lot many things on basis of smaple test. So we are lible to make mistake or commit errors which are given below:

<b>1. Type 1 Error</b> - This type of error arises if we reject the Null hypothesis when it is true which is denoted by $\alpha$.

<b>2. Type 2 Error</b> - Accepting the Null hypothesis when it is false which is denoted by $\beta$.

Note: Type 1 Error is more serious error.

In solving the problem we will fix $ \alpha$ at level of  0.01 or 0.05 (Depends on problem) and then try to minimize $ \beta$
